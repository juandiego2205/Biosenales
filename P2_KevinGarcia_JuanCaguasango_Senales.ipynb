####*Importación de Librerias*






````python
import matplotlib.pyplot as plt
import numpy as np
````

####*Creación de la función Impulso unitario*

````python
def impseq(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]
impulso=impseq(0,-5,5)
plt.stem(impulso[1], impulso[0])
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title(" impulso unitario")
plt.grid()
plt.show()
````

###**PUNTO 1**
#### *Creación de Función Escalón Unitario*




````python
def escalon(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) >= 0
 return [x,n]
esc=escalon(0,-5,5)
plt.stem(esc[1], esc[0])
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Escalon")
plt.grid()
plt.show()
````

###**PUNTO 2**
####*Creación de Función Rampa*



````python
def rampa(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) >= 0
 x1= x*(n-n0)
 return [x1,n]
lista_ram=rampa(-5,-5,1)
plt.stem(lista_ram[1],lista_ram[0] )
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Rampa")
plt.grid()
plt.show()
````

````python
def funcionX(n1,n2):
  n = np.arange(n1,n2+1)
  if len(n)==23:
      P1=rampa(n1,n1,n1+5)[0]
      P2=rampa(n1+6,n1+6,n1+10)[0]
      v=P2[::-1]
      P3=(rampa(n1+11,n1+11,n1+15)[0])+1
      P4=(escalon(n1+16,n1+16,n1+18)[0])*5
      P5=(escalon(n1+19,n1+19,n1+22)[0])*10
      Xn=np.concatenate((P1,v,P3,P4,P5))
      return Xn,n
Xn,n=funcionX(-5,17)
plt.stem(n,Xn)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Sumatoria Rampa")
plt.grid()
plt.show()
````

####**A)** *Creación de la Función*
#####𝑥5(𝑛) = 2𝑥(𝑛 − 4) + 𝑥(𝑛)

````python
X51,n51=funcionX(-1,21)
X52,n52=funcionX(-5,17)
X5=2*X51+X52
def combinacion(x1,n1,x2,n2,pi=0,Pf=0):
  if pi ==0 or Pf==0:
    n = np.arange(min(n1[0], n2[0]), max(n1[-1], n2[-1]) + 1)
  else:
    n = np.arange(pi, Pf + 1)

# Alinear las señales en el mismo rango de tiempo
  X1N = np.zeros(len(n))
  X2N = np.zeros(len(n))

# Llenar los valores correspondientes
  X1N[(n >= n1[0]) & (n <= n1[-1])] = x1
  X2N[(n >= n2[0]) & (n <= n2[-1])] = x2
  return X1N,X2N,n
X51n,X52n,n5=combinacion(X51,n51,X52,n52)

# Sumar las señales alineadas
X5 = 2 * X51n + X52n
plt.stem(n5, X5, use_line_collection=True)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Ecuación Punta A")
plt.grid()
plt.show()
````
