```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def impseq(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]
lista=impseq(1,-1,3)
print(lista[0],lista[1])
plt.stem(lista[1], lista[0])
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Sumatoria de impulsos")
plt.grid()
plt.show()
```

```python
def escalon(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) >= 0

 return [x,n]
lista_esc=escalon(1,-1,3)
plt.stem(lista_esc[1], lista_esc[0])
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Escalon")
plt.grid()
plt.show()
```

##PUNTO 2


```python
def rampa(n0, n1, n2):
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) >= 0
 x1= x*(n-n0)

 return [x1,n]

lista_ram=rampa(1,-4,4)
plt.stem(lista_ram[1],lista_ram[0] )
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Rampa")
plt.grid()
plt.show()
```
